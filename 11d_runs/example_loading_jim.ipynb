# Example loading jim files

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import corner
```

```python
jim_out_file = "./working_1/jim/results_production.npz"
data = np.load(jim_out_file)

chains, log_prob = data["chains"], data["log_prob"]

print(np.shape(chains))
chains = np.reshape(chains, (np.shape(chains)[0]*np.shape(chains)[1], np.shape(chains)[2]))
print(np.shape(chains))

# Thin a bit
jump = 30
chains = chains[::jump]
print("After thinning:")
print(np.shape(chains))

# Sanity check if reshape was done OK
corner.corner(chains)
plt.show()
```
